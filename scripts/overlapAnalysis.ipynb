```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("soft_pf0.1_sp100_lp100_reduced10.overlaps",delim_whitespace=True)
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,5))
df.plot(x="time",y="n_overlaps",ax=ax)
ax.set_ylabel("Number of overlaps",fontsize=16)
ax.set_xlabel(r"Sim time ($\tau$)",fontsize=16)
ax.legend([])
ax.grid(True,linestyle='--')
```
