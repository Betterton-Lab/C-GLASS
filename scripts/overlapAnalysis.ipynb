```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("soft_pf0.4_sp015_lp050_reload003_reduced10.overlaps",delim_whitespace=True)
df['n_tangled'] = df['n_total_crossings_init']-df['n_total_crossings_complete']
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,5))
df.plot(x="time",y="n_instant_bond_overlaps",ax=ax)
ax.set_ylabel("Number of overlaps",fontsize=16)
ax.set_xlabel(r"Sim time ($\tau$)",fontsize=16)
ax.legend([])
ax.grid(True,linestyle='--')
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,5))
df.plot(x="time",y="n_total_crossings_init",ax=ax)
ax.set_ylabel("Number of overlaps",fontsize=16)
ax.set_xlabel(r"Sim time ($\tau$)",fontsize=16)
ax.legend([])
ax.grid(True,linestyle='--')
```

```python
fig,ax = plt.subplots(1,1,figsize=(7,5))
df.plot(x="time",y="n_tangled",ax=ax)
ax.set_ylabel("Number of overlaps",fontsize=16)
ax.set_xlabel(r"Sim time ($\tau$)",fontsize=16)
ax.legend([])
ax.grid(True,linestyle='--')
```
